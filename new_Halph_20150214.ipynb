```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from PIL import Image
import numpy as np
import os
path="/home/tzy/Desktop/python_text/Hα_20150214/"  #待读取的文件夹
path_list=os.listdir(path)
path_list.sort() #对读取的路径进行排序
time_list=[]
for filename in path_list:
    a=os.path.join(path,filename)
    b=a[-10:-3]
    hours=int(b[0:2])-12
    minutes=int(b[2:4])
    seconds=int(b[4:6])
    time=hours*60+minutes+seconds/60
    time_list.append(time)
```

```python
time_list
```

```python
data_sum_H=[]
exptime_H=[]
data_max=[]
path="/home/tzy/Desktop/python_text/Hα_20150214/"  #待读取的文件夹
for filename in path_list:
    print(filename)
    hdu= fits.open(os.path.join(path,filename))
    a=hdu[0].data[1300:1500,1100:1350].sum()
    m=hdu[0].header['EXP_TIME']
    n=hdu[0].header['DATAMAX']
    data_sum_H.append(a)
    data_max.append(n)
    exptime_H.append(m)
```

```python
fig= plt.figure(figsize=(15, 6))

list_time_H=[]
for j in range(len(data_sum_H)):
    list_time_H.append(data_sum_H[j]/exptime_H[j])
#plt.ylim(-2.4*10**(6),2.4*10**(6))
plt.plot(time_list,list_time_H,'y')
#plt.savefig('/home/tzy/Desktop/python_text/picture_20151231/94A.png', format='png', bbox_inches='tight', transparent=True, dpi=600)
```

```python
fig,ax=plt.subplots(figsize=(15,6))
ax.step(time_list,list_time_H)
```
